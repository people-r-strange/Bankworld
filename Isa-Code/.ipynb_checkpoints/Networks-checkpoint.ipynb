```python
import pandas as pd
import networkx as nx
```

```python
data  = pd.read_csv("C:/Users/igome/Documents/Smith/Spring_2021/visual analyticsd/port_6667_0406.csv")
```

```python
#Prepping data for Gephi
```

```python
source = data.groupby(["sourceIP"], sort=False, as_index=False)["time"].first()
```

```python
source['Label'] = source['sourceIP']
source['IP'] = "source" 
```

```python
source
```

```python
dest = data.groupby(["destIP"], sort=False, as_index=False)["time"].first()
```

```python
dest['Label'] = dest['destIP']
dest['IP'] = 'destination'
```

```python
nodes = pd.concat([source,dest], ignore_index = True)
```

```python
nodes['ID'] = nodes.index + 1
```

```python
nodes
```

```python
nodes_list = nodes[['ID','Label','IP' ]]
```

```python
nodes_list.to_csv("C:/Users/igome/Documents/Smith/Spring_2021/visual analyticsd/nodes.csv")
```

```python
#Creating Edges List
```

```python
edge = data[['sourceIP', 'destIP']]
```

```python
nodes
```

```python
edge
```

```python
source = pd.merge(edge, nodes, on = "sourceIP", how = "left")
```

```python
source['destIP'] = source['destIP_x']
source['Source'] = source['ID']
```

```python
target = pd.merge(nodes, edge,on = "destIP", how = "left")
```

```python
target
```

```python
target = target.dropna(subset = ['destIP','ID']).reset_index()
```

```python
target
```

```python
target['Target'] = target['ID']
```

```python
edges = pd.DataFrame()
```

```python
edges['Source'] = source['Source']
edges['Target'] = target['Target']
```

```python
edges
```

```python
edges = edges.groupby(['Source', 'Target']).size().reset_index()
```

```python
edges['Weight'] = edges[0]
```

```python
edges.sort_values("Weight", ascending = False)
```

```python
edges.to_csv("C:/Users/igome/Documents/Smith/Spring_2021/visual analyticsd/edges1.csv")
```
